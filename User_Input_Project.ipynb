```python
import sys
```

1. User name

```python
try:
    Name = str(input("User:"))
    if not Name.isalpha():
        print("Error: Name must contain only alphabetic letters.")
        sys.exit(1)
except Exception as e:
    print(f"An error occurred: {e}")
```

2. User age

```python
try:
    Age_str = input("Age?:")
    if not Age_str.isdigit():
        print("Error: Age must contain only numeric characters.")
        sys.exit(1)
    Age = int(Age_str)
    if Age <= 0 or Age > 122:
        print("Error: Age must be a positive integer and not exceed 122.")
        sys.exit(1)
except Exception as e:
    print(f"An error occurred: {e}")
```

3. User color

```python
try:
  Color= input("Color?:")
  if not Color.isalpha():
    print("Error: Must only contain letters")
    sys.exit(1)
except Exception as e:
  print (f"An error occurred: {e}")
```

4. User password

```python

try:
    Password = input("Password?:")
    if len(Password) < 8:
        print("Error: Password must be at least 8 characters.")
        sys.exit(1)

    has_letter = False
    has_digit = False
    for char in Password:
        if char.isalpha():
            has_letter = True
        elif char.isdigit():
            has_digit = True

    if not (has_letter and has_digit):
        print("Error: Password must contain both letters and digits.")
        sys.exit(1)

    hidden_password = '*' * len(Password)
    print(f"Password (hidden): {hidden_password}")

except Exception as e:
    print(f"An error occurred: {e}")
```

Display


```python
print(f"Your name is {Name}")
print(f"Your age is {Age}")
print(f"The color that you entered is the color {Color}")
print(f"Your password is {hidden_password}")
print("Thank you for providing all of the information correctly!")
```
