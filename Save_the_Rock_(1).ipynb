## Project: Save the Rock!

##### Define functions

```python
#Equations
def find_slope(x1,y1,x2,y2):
    slope = (y2-y1)/(x2-x1)
    return slope
def find_yintercept(x2,y2,slope):
    y_intercept = y2-(slope*x2)
    return y_intercept
def find_xintercept(x2,y2,slope):
    x_intercept = -(y2-(slope*x2))/slope
    return x_intercept


```

##### Defining location

```python
def location (x1,y1,x2,y2,slope):
    slope=find_slope(x1,y1,x2,y2)
    y_intercept=find_yintercept(x2,y2,slope)
    x_intercept=find_xintercept(x2,y2,slope)

    print(f"The Rock is located at point ({x1}, {y1}) and can see point ({x2}, {y2}).\n")
    print(f"The Rock's line has a slope of: {slope}")
    print(f"He can escape through the y-intercept located at y = {y_intercept}")
    print(f"He can escape through the x-intercept located at x = {x_intercept}\n")
    return location
```

```python
x1=4
y1=5
x2=5
y2=6

location(x1,y1,x2,y2,slope)
```
